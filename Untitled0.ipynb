```python
import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer
from sklearn.model_selection import train_test_split

# تحديد مسار ملف البيانات
data_path = "/content/train.csv"

# قراءة البيانات من ملف CSV
df = pd.read_csv(data_path)

# التأكد من أن البيانات تحتوي على عمود 'text'
if 'text' not in df.columns:
    raise ValueError("ملف CSV يجب أن يحتوي على عمود 'text'.")

# تقسيم البيانات إلى تدريب وتحقق بنسبة 90% تدريب و10% تحقق
train_df, validation_df = train_test_split(df, test_size=0.1, random_state=42)

# تحويل DataFrame إلى Dataset
train_dataset = Dataset.from_pandas(train_df)
validation_dataset = Dataset.from_pandas(validation_df)

# تحميل tokenizer
model_path = "/atherLlama-model/model"  # مسار النموذج الذي تستخدمه
tokenizer = AutoTokenizer.from_pretrained(model_path)

# تعيين eos_token كـ pad_token
tokenizer.pad_token = tokenizer.eos_token

# تحديد عمود النص
column_name = "text"

# ترميز البيانات
def tokenize_function(examples):
    return tokenizer(
        examples[column_name],
        padding='max_length',
        truncation=True,
        max_length=512,
    )

print("جاري ترميز البيانات...")

# ترميز بيانات التدريب والتحقق
tokenized_train_dataset = train_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=['text']
)

tokenized_validation_dataset = validation_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=['text']
)

print("تم ترميز البيانات بنجاح.")
```

```python
# إعداد Data Collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,
)

# إعداد معلمات التدريب
training_args = TrainingArguments(
    output_dir="trained_model",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-4,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    save_total_limit=1,
    fp16=True,
    optim="adamw_torch",
    gradient_accumulation_steps=4,
    gradient_checkpointing=True,
    report_to="none",
    dataloader_num_workers=4,
    dataloader_pin_memory=True,
)

# إعداد Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,  # استخدام مجموعة بيانات التدريب
    eval_dataset=tokenized_validation_dataset,  # استخدام مجموعة بيانات التحقق
    data_collator=data_collator,
)

# بدء عملية التدريب
print("بدء عملية التدريب...")
trainer.train()
print("تم انتهاء عملية التدريب.")
```

```python
# حفظ النموذج المدرب
print("جاري حفظ النموذج المدرب...")
trainer.save_model("trained_model")
tokenizer.save_pretrained("trained_model")
print("تم حفظ النموذج المدرب في المجلد 'trained_model'.")
```
